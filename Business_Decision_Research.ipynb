## Convert Target Variable to Numeric

**Reasoning**:
The previous logistic regression model failed due to an 'Unknown label type' error, indicating that the target variable `y` (is_churn) was likely still in a boolean format, which is not suitable for classification models like Logistic Regression. To fix this, I will convert the `is_churn` column to a numerical format (0 for False, 1 for True), then redefine `y` with this numerical column, and finally re-split the data into training and testing sets to ensure the model receives appropriate input.



```python
df['is_churn'] = df['is_churn'].astype(int)

df['Year_Diff']=df['Year_Last_Transaction']-df['Year_First_Transaction']
feature_columns = ['Average_Transaction_Amount', 'Count_Transaction', 'Year_Diff']

X = df[feature_columns]
y = df['is_churn']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

# Inisiasi model logreg
logreg = LogisticRegression()

# fit the model with data
logreg.fit(X_train, y_train)

# Predict model
y_pred = logreg.predict(X_test)

# Evaluasi model menggunakan confusion matrix
cnf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:\n', cnf_matrix)
```
